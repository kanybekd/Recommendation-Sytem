```python
mychoice=ratings.iloc[1].dropna()
mychoice
```

```python
simcandidates=pd.Series()
for i in range(len(mychoice.index)):
    print('adding similars for' + mychoice.index[i] +'...')
    #fetch similar courses with mine
    sims=corrmatrix[mychoice.index[i]].dropna()
    sims=sims.map(lambda x: x*mychoice[i])
    simcandidates=simcandidates.append(sims)
print('sorting......')
simcandidates.sort_values(inplace=True, ascending=False)
print(simcandidates)
```

```python
simcandidates=simcandidates.groupby(simcandidates.index).sum()
```

```python
simcandidates.sort_values(inplace=True, ascending=False)
simcandidates
```
